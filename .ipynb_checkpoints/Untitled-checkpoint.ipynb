```python
import csv

file = "datasetDaLAJsplit.csv"
df = pd.read_csv(file)

pairs = []

with open(file, "r") as f:
    reader = csv.reader(f, delimiter="\t")
    for i, line in enumerate(reader):
        l = line[0].split(',')
        print(l[6])
        if l[6] == 'S-FinV':
            pairs.append((l[1], l[2]))
            
print(pairs)
print(len(pairs))
```
